```python
def clean_df(df,cohort,scores):
    for score in scores:
        df.loc[(df['Cohort Name']==cohort)&(df[score].isna()),score]=df[df['Cohort Name']==cohort][score].mean()
    return df
```

```python
df.isnull().sum()
```

```python
df=clean_df(df,'Cohort1',['Soft skills score','Soft skills pct','Assessment score','Lab score'])
```

```python
df.isnull().sum()
```

```python
df['Final project pct']=df['Final project pct']*100
df['Final project pct']=df['Final project pct'].round(2)
df['Soft skills pct']=df['Soft skills pct']*100
df['Soft skills pct']=df['Soft skills pct'].round(2)
df['Job ready pct']=df['Job ready pct']*100
df['Job ready pct']=df['Job ready pct'].astype(int)
```
